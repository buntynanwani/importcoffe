```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime

# Configuración de estilo
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (15, 10)
plt.rcParams['font.size'] = 10

# Cargar datos
df = pd.read_csv('nasa_power_madrid_completo_2020_2024.csv')
df['Fecha'] = pd.to_datetime(df['Fecha'])

# Crear figura con múltiples subgráficos
fig = plt.figure(figsize=(18, 12))
fig.suptitle('📊 ANÁLISIS DE NUBOSIDAD EN MADRID (2020-2024)', 
             fontsize=20, fontweight='bold', y=0.995)

# ============================================================
# GRÁFICO 1: Serie Temporal Completa
# ============================================================
ax1 = plt.subplot(3, 3, 1)
ax1.plot(df['Fecha'], df['Nubosidad_pct'], linewidth=0.5, alpha=0.7, color='steelblue')
ax1.set_title('☁️ Evolución Temporal de Nubosidad', fontsize=12, fontweight='bold')
ax1.set_xlabel('Fecha')
ax1.set_ylabel('Nubosidad (%)')
ax1.grid(True, alpha=0.3)
ax1.axhline(y=df['Nubosidad_pct'].mean(), color='red', linestyle='--', 
            label=f'Media: {df["Nubosidad_pct"].mean():.1f}%', linewidth=2)
ax1.legend()

# ============================================================
# GRÁFICO 2: Media Mensual
# ============================================================
ax2 = plt.subplot(3, 3, 2)
monthly_avg = df.groupby(['Año', 'Mes'])['Nubosidad_pct'].mean().reset_index()

# 💡 SOLUCIÓN: Renombrar las columnas 'Año' y 'Mes' a 'year' y 'month' para pd.to_datetime
monthly_avg['Fecha_Mes'] = pd.to_datetime(
    monthly_avg[['Año', 'Mes']]
    .rename(columns={'Año': 'year', 'Mes': 'month'})  # <-- ¡CAMBIO AQUÍ!
    .assign(day=1)
)

for year in df['Año'].unique():
    year_data = monthly_avg[monthly_avg['Año'] == year]
    ax2.plot(year_data['Mes'], year_data['Nubosidad_pct'], 
             marker='o', label=str(year), linewidth=2)

ax2.set_title('📅 Nubosidad Media Mensual por Año', fontsize=12, fontweight='bold')
ax2.set_xlabel('Mes')
ax2.set_ylabel('Nubosidad Media (%)')
ax2.set_xticks(range(1, 13))
ax2.set_xticklabels(['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 
                      'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
ax2.legend()
ax2.grid(True, alpha=0.3)

# ============================================================
# GRÁFICO 3: Distribución General
# ============================================================
ax3 = plt.subplot(3, 3, 3)
ax3.hist(df['Nubosidad_pct'], bins=50, color='skyblue', edgecolor='black', alpha=0.7)
ax3.axvline(df['Nubosidad_pct'].mean(), color='red', linestyle='--', 
            linewidth=2, label=f'Media: {df["Nubosidad_pct"].mean():.1f}%')
ax3.axvline(df['Nubosidad_pct'].median(), color='green', linestyle='--', 
            linewidth=2, label=f'Mediana: {df["Nubosidad_pct"].median():.1f}%')
ax3.set_title('📈 Distribución de Nubosidad', fontsize=12, fontweight='bold')
ax3.set_xlabel('Nubosidad (%)')
ax3.set_ylabel('Frecuencia (días)')
ax3.legend()
ax3.grid(True, alpha=0.3)

# ============================================================
# GRÁFICO 4: Box Plot por Año
# ============================================================
ax4 = plt.subplot(3, 3, 4)
df.boxplot(column='Nubosidad_pct', by='Año', ax=ax4)
ax4.set_title('📦 Distribución de Nubosidad por Año', fontsize=12, fontweight='bold')
ax4.set_xlabel('Año')
ax4.set_ylabel('Nubosidad (%)')
plt.suptitle('')  # Eliminar título automático del boxplot

# ============================================================
# GRÁFICO 5: Box Plot por Estación
# ============================================================
ax5 = plt.subplot(3, 3, 5)
estaciones_orden = ['Invierno', 'Primavera', 'Verano', 'Otoño']
df['Estacion'] = pd.Categorical(df['Estacion'], categories=estaciones_orden, ordered=True)
df.boxplot(column='Nubosidad_pct', by='Estacion', ax=ax5)
ax5.set_title('🌦️ Distribución de Nubosidad por Estación', fontsize=12, fontweight='bold')
ax5.set_xlabel('Estación')
ax5.set_ylabel('Nubosidad (%)')
plt.suptitle('')

# ============================================================
# GRÁFICO 6: Heatmap Año-Mes
# ============================================================
ax6 = plt.subplot(3, 3, 6)
pivot_data = df.pivot_table(values='Nubosidad_pct', 
                             index='Mes', 
                             columns='Año', 
                             aggfunc='mean')
sns.heatmap(pivot_data, annot=True, fmt='.1f', cmap='RdYlBu_r', 
            ax=ax6, cbar_kws={'label': 'Nubosidad (%)'})
ax6.set_title('🗓️ Mapa de Calor: Nubosidad Promedio', fontsize=12, fontweight='bold')
ax6.set_ylabel('Mes')
ax6.set_yticklabels(['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 
                     'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'], rotation=0)

# ============================================================
# GRÁFICO 7: Días Despejados vs Nublados por Año
# ============================================================
ax7 = plt.subplot(3, 3, 7)
categorias = []
for _, row in df.iterrows():
    if row['Nubosidad_pct'] < 25:
        categorias.append('Despejado')
    elif row['Nubosidad_pct'] < 50:
        categorias.append('Poco Nublado')
    elif row['Nubosidad_pct'] < 75:
        categorias.append('Nublado')
    else:
        categorias.append('Muy Nublado')

df['Categoria_Cielo'] = categorias

conteo_categorias = df.groupby(['Año', 'Categoria_Cielo']).size().unstack(fill_value=0)
conteo_categorias.plot(kind='bar', stacked=True, ax=ax7, 
                       color=['gold', 'lightskyblue', 'lightgray', 'darkgray'])
ax7.set_title('☀️ Categorías de Cielo por Año', fontsize=12, fontweight='bold')
ax7.set_xlabel('Año')
ax7.set_ylabel('Número de Días')
ax7.legend(title='Categoría')
ax7.set_xticklabels(ax7.get_xticklabels(), rotation=0)

# ============================================================
# GRÁFICO 8: Tendencia Promedio Anual
# ============================================================
ax8 = plt.subplot(3, 3, 8)
yearly_avg = df.groupby('Año')['Nubosidad_pct'].mean()
ax8.bar(yearly_avg.index, yearly_avg.values, color='steelblue', edgecolor='black')
ax8.plot(yearly_avg.index, yearly_avg.values, marker='o', color='red', 
         linewidth=2, markersize=8)
ax8.set_title('📊 Promedio Anual de Nubosidad', fontsize=12, fontweight='bold')
ax8.set_xlabel('Año')
ax8.set_ylabel('Nubosidad Media (%)')
ax8.grid(True, alpha=0.3, axis='y')

# Agregar valores en las barras
for i, v in enumerate(yearly_avg.values):
    ax8.text(yearly_avg.index[i], v + 1, f'{v:.1f}%', 
             ha='center', fontweight='bold')

# ============================================================
# GRÁFICO 9: Promedio por Estación (todos los años)
# ============================================================
ax9 = plt.subplot(3, 3, 9)
season_avg = df.groupby('Estacion')['Nubosidad_pct'].mean().reindex(estaciones_orden)
colors = ['lightblue', 'lightgreen', 'gold', 'orange']
bars = ax9.bar(season_avg.index, season_avg.values, color=colors, edgecolor='black')
ax9.set_title('🍂 Nubosidad Media por Estación', fontsize=12, fontweight='bold')
ax9.set_xlabel('Estación')
ax9.set_ylabel('Nubosidad Media (%)')
ax9.grid(True, alpha=0.3, axis='y')

# Agregar valores en las barras
for bar in bars:
    height = bar.get_height()
    ax9.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{height:.1f}%', ha='center', fontweight='bold')

plt.tight_layout()
plt.savefig('analisis_nubosidad_madrid.png', dpi=300, bbox_inches='tight')
print("✅ Gráfico guardado como 'analisis_nubosidad_madrid.png'")
plt.show()

# ============================================================
# ESTADÍSTICAS DETALLADAS
# ============================================================
print("\n" + "="*70)
print("📊 ESTADÍSTICAS DETALLADAS DE NUBOSIDAD - MADRID (2020-2024)")
print("="*70)

print("\n📈 ESTADÍSTICAS GENERALES:")
print(f"   Media general: {df['Nubosidad_pct'].mean():.2f}%")
print(f"   Mediana: {df['Nubosidad_pct'].median():.2f}%")
print(f"   Desviación estándar: {df['Nubosidad_pct'].std():.2f}%")
print(f"   Mínimo: {df['Nubosidad_pct'].min():.2f}%")
print(f"   Máximo: {df['Nubosidad_pct'].max():.2f}%")

print("\n📅 ESTADÍSTICAS POR AÑO:")
for year in sorted(df['Año'].unique()):
    year_data = df[df['Año'] == year]['Nubosidad_pct']
    print(f"   {year}: Media = {year_data.mean():.2f}%, "
          f"Mín = {year_data.min():.2f}%, "
          f"Máx = {year_data.max():.2f}%")

print("\n🌦️ ESTADÍSTICAS POR ESTACIÓN:")
for estacion in estaciones_orden:
    season_data = df[df['Estacion'] == estacion]['Nubosidad_pct']
    print(f"   {estacion}: Media = {season_data.mean():.2f}%, "
          f"Mín = {season_data.min():.2f}%, "
          f"Máx = {season_data.max():.2f}%")

print("\n☀️ CATEGORÍAS DE CIELO (TODOS LOS AÑOS):")
total_dias = len(df)
for categoria in ['Despejado', 'Poco Nublado', 'Nublado', 'Muy Nublado']:
    dias = len(df[df['Categoria_Cielo'] == categoria])
    porcentaje = (dias / total_dias) * 100
    print(f"   {categoria}: {dias} días ({porcentaje:.1f}%)")

print("\n🎯 MESES CON MAYOR Y MENOR NUBOSIDAD:")
monthly_avg_all = df.groupby('Mes')['Nubosidad_pct'].mean()
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
         'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
mes_max = monthly_avg_all.idxmax()
mes_min = monthly_avg_all.idxmin()
print(f"   Mayor nubosidad: {meses[mes_max-1]} ({monthly_avg_all[mes_max]:.2f}%)")
print(f"   Menor nubosidad: {meses[mes_min-1]} ({monthly_avg_all[mes_min]:.2f}%)")

print("\n💡 IMPLICACIONES PARA ENERGÍA SOLAR:")
dias_optimos = len(df[df['Nubosidad_pct'] < 30])
dias_buenos = len(df[(df['Nubosidad_pct'] >= 30) & (df['Nubosidad_pct'] < 60)])
dias_malos = len(df[df['Nubosidad_pct'] >= 60])
print(f"   Días óptimos para solar (<30% nubosidad): {dias_optimos} ({dias_optimos/total_dias*100:.1f}%)")
print(f"   Días buenos (30-60% nubosidad): {dias_buenos} ({dias_buenos/total_dias*100:.1f}%)")
print(f"   Días desfavorables (>60% nubosidad): {dias_malos} ({dias_malos/total_dias*100:.1f}%)")

print("\n" + "="*70)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# ============================================================
# PARÁMETROS DEL SISTEMA FOTOVOLTAICO
# ============================================================
class SistemaFotovoltaico:
    """
    Parámetros típicos de un sistema solar fotovoltaico
    """
    def __init__(self, area_m2=100, eficiencia=0.18, performance_ratio=0.75):
        """
        Args:
            area_m2: Área de paneles solares en m² (default: 100m²)
            eficiencia: Eficiencia de conversión del panel (default: 18% = 0.18)
                       - Paneles económicos: 15-17%
                       - Paneles estándar: 18-20%
                       - Paneles premium: 20-23%
            performance_ratio: Ratio de rendimiento del sistema (default: 75% = 0.75)
                              Incluye pérdidas por:
                              - Temperatura (5-10%)
                              - Cableado (2-3%)
                              - Inversor (3-5%)
                              - Suciedad/polvo (2-5%)
                              - Sombreado (0-5%)
        """
        self.area_m2 = area_m2
        self.eficiencia = eficiencia
        self.performance_ratio = performance_ratio
        self.potencia_pico_kWp = (area_m2 * 1) * (eficiencia)  # 1 kW/m² radiación estándar

# ============================================================
# FÓRMULA DE CÁLCULO
# ============================================================
"""
ENERGÍA GENERADA (kWh) = 
    Radiación Solar (kWh/m²) × 
    Área de Paneles (m²) × 
    Eficiencia del Panel (%) × 
    Performance Ratio (%)

Ejemplo con tus datos:
- Radiación: 5.2 kWh/m²/día
- Área: 100 m²
- Eficiencia: 18%
- PR: 75%

Energía = 5.2 × 100 × 0.18 × 0.75 = 70.2 kWh/día
"""

# ============================================================
# CARGAR Y PREPARAR DATOS
# ============================================================
print("=" * 80)
print("⚡ CALCULADORA DE GENERACIÓN DE ENERGÍA SOLAR - MADRID")
print("=" * 80)

df = pd.read_csv('nasa_power_madrid_completo_2020_2024.csv')
df['Fecha'] = pd.to_datetime(df['Fecha'])

# Crear instancia del sistema (puedes modificar estos valores)
sistema = SistemaFotovoltaico(
    area_m2=100,        # 100 m² de paneles (típico para vivienda o pequeño edificio)
    eficiencia=0.18,    # 18% eficiencia
    performance_ratio=0.75  # 75% PR
)

print(f"\n🔧 CONFIGURACIÓN DEL SISTEMA:")
print(f"   📐 Área de paneles: {sistema.area_m2} m²")
print(f"   ⚙️  Eficiencia: {sistema.eficiencia*100:.0f}%")
print(f"   📊 Performance Ratio: {sistema.performance_ratio*100:.0f}%")
print(f"   ⚡ Potencia Pico: {sistema.potencia_pico_kWp:.2f} kWp")

# ============================================================
# CÁLCULO DE ENERGÍA GENERADA
# ============================================================

# Energía diaria (kWh)
df['Energia_Generada_kWh_dia'] = (
    df['Radiacion_Solar_Global_kWh_m2'] * 
    sistema.area_m2 * 
    sistema.eficiencia * 
    sistema.performance_ratio
)

# Ajuste por temperatura (los paneles pierden eficiencia con calor)
# Por cada grado sobre 25°C, se pierde ~0.4% de eficiencia
TEMP_REFERENCIA = 25  # °C
COEF_TEMP = -0.004  # -0.4% por °C

df['Ajuste_Temperatura'] = 1 + (COEF_TEMP * (df['Temperatura_Media_C'] - TEMP_REFERENCIA))
df['Energia_Ajustada_Temp_kWh_dia'] = df['Energia_Generada_kWh_dia'] * df['Ajuste_Temperatura']

# Ajuste por nubosidad (reducción adicional)
# Aunque la radiación ya considera nubosidad, añadimos factor de seguridad
df['Factor_Nubosidad'] = 1 - (df['Nubosidad_pct'] / 100) * 0.2  # 20% impacto máximo
df['Energia_Final_kWh_dia'] = df['Energia_Ajustada_Temp_kWh_dia'] * df['Factor_Nubosidad']

# Energía mensual y anual
df['Energia_Mensual_kWh'] = df.groupby([df['Fecha'].dt.year, df['Fecha'].dt.month])['Energia_Final_kWh_dia'].transform('sum')
df['Energia_Anual_kWh'] = df.groupby(df['Fecha'].dt.year)['Energia_Final_kWh_dia'].transform('sum')

# ============================================================
# ESTADÍSTICAS
# ============================================================
print(f"\n📈 RESULTADOS DE GENERACIÓN:")
print(f"   ☀️  Energía diaria promedio: {df['Energia_Final_kWh_dia'].mean():.2f} kWh/día")
print(f"   📅 Energía mensual promedio: {df['Energia_Mensual_kWh'].mean():.2f} kWh/mes")
print(f"   📆 Energía anual promedio: {df['Energia_Anual_kWh'].mean():.2f} kWh/año")
print(f"   ⬆️  Día con mayor generación: {df['Energia_Final_kWh_dia'].max():.2f} kWh")
print(f"   ⬇️  Día con menor generación: {df['Energia_Final_kWh_dia'].min():.2f} kWh")

# Comparación con consumo de hogares
CONSUMO_HOGAR_ANUAL = 3500  # kWh/año (promedio España)
hogares_abastecidos = df['Energia_Anual_kWh'].mean() / CONSUMO_HOGAR_ANUAL
print(f"\n🏠 EQUIVALENCIAS:")
print(f"   Hogares que puede abastecer: {hogares_abastecidos:.1f}")
print(f"   Consumo mensual cubierto: {(df['Energia_Mensual_kWh'].mean() / (CONSUMO_HOGAR_ANUAL/12)) * 100:.1f}%")

# Análisis por estación
print(f"\n🌦️  GENERACIÓN POR ESTACIÓN:")
estaciones_orden = ['Invierno', 'Primavera', 'Verano', 'Otoño']
for estacion in estaciones_orden:
    energia_estacion = df[df['Estacion'] == estacion]['Energia_Final_kWh_dia'].mean()
    print(f"   {estacion}: {energia_estacion:.2f} kWh/día")

# Análisis por año
print(f"\n📊 GENERACIÓN POR AÑO:")
for year in sorted(df['Año'].unique()):
    year_data = df[df['Año'] == year]
    energia_anual = year_data['Energia_Final_kWh_dia'].sum()
    print(f"   {year}: {energia_anual:.0f} kWh/año ({energia_anual/12:.0f} kWh/mes)")

# ============================================================
# ANÁLISIS ECONÓMICO
# ============================================================
print(f"\n💰 ANÁLISIS ECONÓMICO:")
PRECIO_kWh = 0.15  # €/kWh (precio medio España)
COSTE_INSTALACION = 6000  # € (para 100m² sistema completo)

ahorro_anual = df['Energia_Anual_kWh'].mean() * PRECIO_kWh
payback_years = COSTE_INSTALACION / ahorro_anual
ahorro_25_años = (ahorro_anual * 25) - COSTE_INSTALACION

print(f"   💶 Ahorro anual: {ahorro_anual:.0f} €/año")
print(f"   📈 Periodo de retorno: {payback_years:.1f} años")
print(f"   🎯 Ahorro total (25 años): {ahorro_25_años:.0f} €")
print(f"   🌍 CO₂ evitado (25 años): {(df['Energia_Anual_kWh'].mean() * 25 * 0.3 / 1000):.1f} toneladas")

# ============================================================
# VISUALIZACIONES
# ============================================================
fig = plt.figure(figsize=(18, 12))
fig.suptitle('⚡ ANÁLISIS DE GENERACIÓN DE ENERGÍA SOLAR - MADRID', 
             fontsize=20, fontweight='bold', y=0.995)

# GRÁFICO 1: Serie temporal de energía generada
ax1 = plt.subplot(3, 3, 1)
ax1.plot(df['Fecha'], df['Energia_Final_kWh_dia'], linewidth=0.5, alpha=0.7, color='orange')
ax1.set_title('☀️ Energía Generada Diaria', fontsize=12, fontweight='bold')
ax1.set_xlabel('Fecha')
ax1.set_ylabel('Energía (kWh/día)')
ax1.axhline(y=df['Energia_Final_kWh_dia'].mean(), color='red', linestyle='--', 
            label=f'Media: {df["Energia_Final_kWh_dia"].mean():.1f} kWh/día', linewidth=2)
ax1.legend()
ax1.grid(True, alpha=0.3)

# GRÁFICO 2: Comparación Radiación vs Energía

ax2 = plt.subplot(3, 3, 2)

# 💡 SOLUCIÓN 1: Asignar nombres explícitos al agrupar
monthly = df.groupby([df['Fecha'].dt.year.rename('Año'), 
                      df['Fecha'].dt.month.rename('Mes')]).agg({
    'Radiacion_Solar_Global_kWh_m2': 'mean',
    'Energia_Final_kWh_dia': 'mean'
}).reset_index()

# La línea de pd.to_datetime ahora debe ajustarse a los nuevos nombres: 'Año' y 'Mes'
monthly['Fecha_Mes'] = pd.to_datetime(
    monthly[['Año', 'Mes']] # <-- Usamos 'Año' y 'Mes'
    .rename(columns={'Año': 'year', 'Mes': 'month'})
    .assign(day=1)
)



ax2_twin = ax2.twinx()
ax2.plot(monthly.index, monthly['Radiacion_Solar_Global_kWh_m2'], 
         color='gold', marker='o', label='Radiación Solar', linewidth=2)
ax2_twin.plot(monthly.index, monthly['Energia_Final_kWh_dia'], 
              color='green', marker='s', label='Energía Generada', linewidth=2)
ax2.set_title('📊 Radiación vs Energía Generada', fontsize=12, fontweight='bold')
ax2.set_xlabel('Mes')
ax2.set_ylabel('Radiación (kWh/m²/día)', color='gold')
ax2_twin.set_ylabel('Energía (kWh/día)', color='green')
ax2.legend(loc='upper left')
ax2_twin.legend(loc='upper right')
ax2.grid(True, alpha=0.3)

# GRÁFICO 3: Distribución de energía generada
ax3 = plt.subplot(3, 3, 3)
ax3.hist(df['Energia_Final_kWh_dia'], bins=50, color='orange', edgecolor='black', alpha=0.7)
ax3.axvline(df['Energia_Final_kWh_dia'].mean(), color='red', linestyle='--', 
            linewidth=2, label=f'Media: {df["Energia_Final_kWh_dia"].mean():.1f} kWh')
ax3.set_title('📈 Distribución de Energía Diaria', fontsize=12, fontweight='bold')
ax3.set_xlabel('Energía (kWh/día)')
ax3.set_ylabel('Frecuencia (días)')
ax3.legend()
ax3.grid(True, alpha=0.3)

# GRÁFICO 4: Energía por estación
ax4 = plt.subplot(3, 3, 4)
season_energy = df.groupby('Estacion')['Energia_Final_kWh_dia'].mean().reindex(estaciones_orden)
colors_season = ['lightblue', 'lightgreen', 'gold', 'orange']
bars = ax4.bar(season_energy.index, season_energy.values, color=colors_season, edgecolor='black')
ax4.set_title('🌦️ Energía Media por Estación', fontsize=12, fontweight='bold')
ax4.set_ylabel('Energía (kWh/día)')
ax4.grid(True, alpha=0.3, axis='y')
for bar in bars:
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{height:.1f}', ha='center', fontweight='bold')

# GRÁFICO 5: Energía anual
ax5 = plt.subplot(3, 3, 5)
yearly_energy = df.groupby('Año')['Energia_Final_kWh_dia'].sum()
ax5.bar(yearly_energy.index, yearly_energy.values, color='green', edgecolor='black')
ax5.plot(yearly_energy.index, yearly_energy.values, marker='o', color='red', linewidth=2)
ax5.set_title('📆 Generación Anual Total', fontsize=12, fontweight='bold')
ax5.set_xlabel('Año')
ax5.set_ylabel('Energía (kWh/año)')
ax5.grid(True, alpha=0.3, axis='y')
for i, v in enumerate(yearly_energy.values):
    ax5.text(yearly_energy.index[i], v + 200, f'{v:.0f}', ha='center', fontweight='bold')

# GRÁFICO 6: Heatmap mensual de energía
ax6 = plt.subplot(3, 3, 6)
pivot_energy = df.pivot_table(values='Energia_Final_kWh_dia', 
                               index='Mes', 
                               columns='Año', 
                               aggfunc='mean')
sns.heatmap(pivot_energy, annot=True, fmt='.1f', cmap='YlOrRd', 
            ax=ax6, cbar_kws={'label': 'Energía (kWh/día)'})
ax6.set_title('🗓️ Mapa de Calor: Energía Promedio', fontsize=12, fontweight='bold')
ax6.set_ylabel('Mes')
ax6.set_yticklabels(['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 
                     'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'], rotation=0)

# GRÁFICO 7: Impacto de temperatura
ax7 = plt.subplot(3, 3, 7)
temp_bins = pd.cut(df['Temperatura_Media_C'], bins=10)
temp_impact = df.groupby(temp_bins)['Energia_Final_kWh_dia'].mean()
temp_labels = [f'{int(i.left)}-{int(i.right)}°C' for i in temp_impact.index]
ax7.plot(range(len(temp_impact)), temp_impact.values, marker='o', color='red', linewidth=2)
ax7.set_title('🌡️ Impacto de Temperatura en Generación', fontsize=12, fontweight='bold')
ax7.set_xlabel('Rango de Temperatura')
ax7.set_ylabel('Energía Media (kWh/día)')
ax7.set_xticks(range(len(temp_impact)))
ax7.set_xticklabels(temp_labels, rotation=45, ha='right')
ax7.grid(True, alpha=0.3)

# GRÁFICO 8: Energía mensual promedio
ax8 = plt.subplot(3, 3, 8)
monthly_avg_energy = df.groupby('Mes')['Energia_Final_kWh_dia'].mean()
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
ax8.bar(monthly_avg_energy.index, monthly_avg_energy.values, color='orange', edgecolor='black')
ax8.set_title('📅 Energía Media Mensual', fontsize=12, fontweight='bold')
ax8.set_xlabel('Mes')
ax8.set_ylabel('Energía (kWh/día)')
ax8.set_xticks(range(1, 13))
ax8.set_xticklabels(meses)
ax8.grid(True, alpha=0.3, axis='y')

# GRÁFICO 9: Acumulado anual
ax9 = plt.subplot(3, 3, 9)
for year in sorted(df['Año'].unique()):
    year_data = df[df['Año'] == year].sort_values('Fecha')
    year_data['Energia_Acumulada'] = year_data['Energia_Final_kWh_dia'].cumsum()
    day_of_year = range(1, len(year_data) + 1)
    ax9.plot(day_of_year, year_data['Energia_Acumulada'], label=str(year), linewidth=2)

ax9.set_title('📈 Energía Acumulada Anual', fontsize=12, fontweight='bold')
ax9.set_xlabel('Día del Año')
ax9.set_ylabel('Energía Acumulada (kWh)')
ax9.legend()
ax9.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('analisis_generacion_solar_madrid.png', dpi=300, bbox_inches='tight')
print("\n✅ Gráficos guardados como 'analisis_generacion_solar_madrid.png'")
plt.show()

# ============================================================
# GUARDAR RESULTADOS
# ============================================================
df_export = df[['Fecha', 'Radiacion_Solar_Global_kWh_m2', 'Temperatura_Media_C', 
                'Nubosidad_pct', 'Energia_Final_kWh_dia', 'Año', 'Mes', 'Estacion']]
df_export.to_csv('energia_solar_generada_madrid.csv', index=False)
print("✅ Datos guardados en 'energia_solar_generada_madrid.csv'")

print("\n" + "=" * 80)
```
